## HOW TO GET **Education Benefits** LICENSING ON GITHUB

To get access to GitHub Education benefits, you usually need to verify that you are a student, teacher, or school administrator. GitHub asks for proof of enrollment or teaching status so it can confirm that you qualify for the program. Once your application is approved, you can use the benefits included in the Education Pack, such as developer tools, learning resources, and other student-focused offers.

The process is handled through GitHub Education, where you submit your school information and any required documents. If your verification is successful, GitHub grants access to the benefits on your account. This can help you learn faster, build projects more easily, and take advantage of resources that support education and software development.

If you are referring to a specific license, product, or feature on GitHub, the exact steps may depend on what you want to activate. In general, though, the key requirement is to prove your education status and complete GitHub’s verification process.

## STEPS

- Open GitHub: https://github.com
- Create a new free account.
- Enter your email, password, username, and country/region.
- Go to **Settings** and open **Billing and licensing**.
- Open **GitHub Education Benefits** to access free developer resources.
- Start a new application.
- Select your role in education: **Student** or **Teacher**.
- Search and select your school/university.
- In "What is the name of your school?", enter your university name.
- Enter your university email address.
- Go to the option to enable **2FA** (two-factor authentication) for account security.
- Select the type of proof you want to provide.
- Choose the first option: **Dated school ID**.
- Enter your university address for ID location details.
- Submit your application.
- Check your email and complete all verification steps.


## Important Notes

- Use real and accurate school information.
- Make sure your document is clear and valid before uploading.
- Keep 2FA enabled to protect your account.
- Approval may take some time depending on verification volume.



## PYTHON STEPS 

```python
# Full browser automation with Python (Selenium).
# Personal fields remain manual (email, password, username, school details, proof upload, submit).

from pathlib import Path
import time
import sys
import subprocess


def ensure_package(pkg_name: str) -> None:
    try:
        __import__(pkg_name)
    except Exception:
        print(f"Installing missing package: {pkg_name}")
        subprocess.check_call([sys.executable, "-m", "pip", "install", pkg_name])


ensure_package("selenium")

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


def wait_for_user(message: str) -> None:
    input(f"\n{message}\nPress Enter to continue...")


shots_dir = Path("github_education_screens_selenium")
shots_dir.mkdir(parents=True, exist_ok=True)


def build_driver():
    # Try Chrome first, then Edge as fallback.
    try:
        opts = webdriver.ChromeOptions()
        opts.add_argument("--start-maximized")
        return webdriver.Chrome(options=opts), "Chrome"
    except Exception:
        opts = webdriver.EdgeOptions()
        opts.add_argument("--start-maximized")
        return webdriver.Edge(options=opts), "Edge"


driver, browser_name = build_driver()
wait = WebDriverWait(driver, 30)
print(f"Using browser: {browser_name}")

try:
    # Step 1: GitHub home
    driver.get("https://github.com/")
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    driver.save_screenshot(str(shots_dir / "01_home.png"))
    print("Step 1 complete: Opened GitHub home.")

    # Step 2: Sign-up page
    driver.get("https://github.com/signup")
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    driver.save_screenshot(str(shots_dir / "02_signup.png"))
    print("Step 2 complete: Opened sign-up page.")

    # Step 3: Manual personal fields
    wait_for_user(
        "Step 3 (manual): Enter email, password, username, and country/region, then create your account."
    )
    driver.save_screenshot(str(shots_dir / "03_after_manual_signup.png"))

    # Step 4: Billing/Licensing
    driver.get("https://github.com/settings/billing")
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    driver.save_screenshot(str(shots_dir / "04_billing.png"))
    print("Step 4 complete: Opened Billing/Licensing settings.")

    # Step 5: Education benefits
    driver.get("https://education.github.com/pack")
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    driver.save_screenshot(str(shots_dir / "05_education_pack.png"))
    print("Step 5 complete: Opened GitHub Education benefits page.")

    # Step 6: Start application
    driver.get("https://education.github.com/discount_requests/application")
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    driver.save_screenshot(str(shots_dir / "06_start_application.png"))
    print("Step 6 complete: Opened Education application page.")

    # Step 7-14: Manual school/application fields + proof upload
    wait_for_user(
        "Step 7-14 (manual): Select role (Student/Teacher), school, school name, university email, address, proof type, and upload dated school ID."
    )
    driver.save_screenshot(str(shots_dir / "07_after_manual_form.png"))

    # Step 15: Manual submit
    wait_for_user("Step 15 (manual): Click Submit on the application page.")
    driver.save_screenshot(str(shots_dir / "08_after_submit.png"))

    # Step 16: 2FA page
    driver.get("https://github.com/settings/security")
    wait.until(EC.presence_of_element_located((By.TAG_NAME, "body")))
    driver.save_screenshot(str(shots_dir / "09_security_2fa.png"))
    print("Step 16 complete: Opened 2FA security page.")

    wait_for_user("Final manual step: Enable 2FA and verify email notifications.")
    driver.save_screenshot(str(shots_dir / "10_after_2fa.png"))

    print("\nAutomation finished.")
    print(f"Screenshots saved in: {shots_dir.resolve()}")

finally:
    try:
        time.sleep(2)
    except Exception:
        pass
    print("Skipping driver.quit() to avoid notebook shutdown errors.")
```
